```python
from ultralytics import YOLO
import cv2
import time

class ObjectDetector:
    def __init__(self, model_path):
        self.model = YOLO(model_path)
        self.class_names = self.model.model.names if hasattr(self.model.model, 'names') else self.model.names
        self.target_class_names = ['person', 'cup', 'bottle', 'book']
        self.target_class_ids = [list(self.class_names.keys())[list(self.class_names.values()).index(name)] for name in self.target_class_names]
        self.CONFIDENCE_THRESHOLD = 0.80
        self.NMS_THRESHOLD = 0.75
        self.INPUT_WIDTH = 640  # Adjust based on your needs
        self.INPUT_HEIGHT = 480

    def detect(self):
        cap = cv2.VideoCapture(0)

        while True:
            start_time = time.time()
            
            # Capture frame from the webcam
            ret, frame = cap.read()
            
            if not ret:
                break
            
            # Resize the frame for faster processing
            resized_frame = cv2.resize(frame, (self.INPUT_WIDTH, self.INPUT_HEIGHT))

            # Run inference on the resized frame
            results = self.model(resized_frame)

            if results:
                for result in results:
                    detections = result.boxes.data.tolist()

                    for detection in detections:
                        x1, y1, x2, y2, confidence, class_id = detection
                        class_label = self.class_names[int(class_id)]

                        if confidence > self.CONFIDENCE_THRESHOLD and int(class_id) in self.target_class_ids:
                            # Draw a rectangle around the detected object
                            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
                            # Put a label with the class name and confidence
                            cv2.putText(frame, f'{class_label} {confidence:.2f}', (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36, 255, 12), 2)
            
            # Calculate and display FPS
            end_time = time.time()
            fps = 1 / (end_time - start_time)
            cv2.putText(frame, f'FPS: {fps:.2f}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

            # Display the resulting frame
            cv2.imshow('frame', frame)

            if cv2.waitKey(25) & 0xFF == ord('q'):
                break

        # When everything is done, release the capture
        cap.release()
        cv2.destroyAllWindows()

# Create an instance of the ObjectDetector class
detector = ObjectDetector('yolov8n.pt')

# Start the object detection
detector.detect()
```

## ONLY PERSON DETECTION


```python
import cv2
import math
from ultralytics import YOLO

# Start webcam
cap = cv2.VideoCapture(0)
cap.set(3, 640)  # Set width
cap.set(4, 480)  # Set height

# Load model
model = YOLO("yolo-Weights/yolov8n.pt")

# Object classes to detect
classNames = ["person", "bottle", "cup", "cell phone"]

while True:
    success, img = cap.read()
    if not success:
        break

    results = model(img, stream=True)

    # Process detection results
    for r in results:
        boxes = r.boxes
        for box in boxes:
            # Extract bounding box coordinates
            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)  # Convert to int values

            # Extract confidence
            confidence = math.ceil((box.conf[0] * 100)) / 100

            # Extract class name
            cls = int(box.cls[0])
            if cls < len(classNames):
                className = classNames[cls]

                # Draw rectangle if class is in the specified classes
                cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 255), 3)

                # Add label
                label = f"{className} {confidence:.4f}"
                org = (x1, y1 - 10)
                font = cv2.FONT_HERSHEY_SIMPLEX
                fontScale = 0.5
                color = (255, 0, 0)
                thickness = 2
                cv2.putText(img, label, org, font, fontScale, color, thickness)

    # Show webcam feed
    cv2.imshow('Webcam', img)
    
    # Exit if 'q' is pressed
    if cv2.waitKey(1) == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2

# Load the YOLOv8 model
model = YOLO('yolov8n.pt')

# Open the webcam
cap = cv2.VideoCapture(0)

while True:
    # Capture frame from the webcam
    ret, frame = cap.read()
    
    # Run inference on the frame
    results = model(frame)  # <sup className="rounded-full text-xs cursor-pointer [&>*]:text-white h-4 w-4 px-1 bg-zinc-400 hover:bg-zinc-500 dark:bg-zinc-700 hover:dark:bg-zinc-600">[1](https://stackoverflow.com/questions/75925797)</sup> 
    
    # Get the annotated frame with bounding boxes
    annotated_frame = results[0].plot()  # <sup className="rounded-full text-xs cursor-pointer [&>*]:text-white h-4 w-4 px-1 bg-zinc-400 hover:bg-zinc-500 dark:bg-zinc-700 hover:dark:bg-zinc-600">[1](https://stackoverflow.com/questions/75925797)</sup> 
    
    # Convert annotated frame to OpenCV format
    annotated_frame = cv2.cvtColor(annotated_frame, cv2.COLOR_RGB2BGR)
    
    # Filter the results to only include the desired classes
    filtered_results = []
    for detection in results[0].boxes:
        if detection.cls in [0, 6, 14]:  # Map class indices to class names
            filtered_results.append(detection)
    
    # Draw bounding boxes for the filtered results
    for detection in filtered_results:
        x1, y1, x2, y2 = detection.xyxy[0]
        cv2.rectangle(annotated_frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
    
    # Display the annotated frame
    cv2.imshow('mat', annotated_frame)
    
    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```
